# Grouping

```python
text = '123-4567 is my telephone.'
```

```python
re.findall(r'[\d]{3}-[\d]{4}',text)
```

```python
re.findall(r'([\d]{3})-([\d]{4})',text)
```

```python
k = re.search(r'([\d]{3})-([\d]{4})',text)
```

```python
k.group()
```

```python
k.groups()
```

```python
k.group(1)
```

```python
k.group(2)
```

```python
?P<group-name>
```

```python
k = re.search(r'(?P<vijay>[\d]{3})-(?P<ajay>[\d]{4})',text)
```

```python
k.group('vijay')
```

```python
k.group('ajay')
```

# For Practise

```python
text = ['123 456 7890','(123) 456 7890','123-456-7890','123.456.7890','1234567890']
```

```python
pat = r'\(?\d{3}\)?[\W]?\d{3}[\W]?\d{4}'
```

```python
for i in text:
    m = re.search(pat,i)
    if m:
        print(m.group())
```

```python
text = ['1 123 456 7890','+1 (123) 456 7890','123-456-7890','123.456.7890','1234567890']
pat = r'\+?\d?\s?\(?\d{3}\)?[\W]?\d{3}[\W]?\d{4}'
for i in text:
    m = re.search(pat,i)
    if m:
        print(m.group())
```

```python
text = ['1 123 456 7890','+1 (123) 456 7890','123-456-7890','123.456.7890','1234567890']
for i in text:
    print(i)
```

```python
text = ['1 123 456 7890','+1 (123) 456 7890','123-456-7890','123.456.7890','1234567890']
pat = r'\+?\d?\s?\(?\d{3}\)?[\W]?\d{3}[\W]?\d{4}'
pact = re.compile(pat)
for i in text:
    m = re.search(pact,i)
    if m:
        print(m.group())
```

```python
text = ['1 123 456 7890','+1 (123) 456 7890','123-456-7890','123.456.7890','1234567890']
pat = r'(\+?\d?)\s?(\(?\d{3}\)?)[\W]?(\d{3})[\W]?(\d{4})'
pact = re.compile(pat)
for i in text:
    m = re.search(pact,i)
    if m:
        print(m.group(),'\t', m.group(1),'\t',m.group(2),'\t',m.group(3),'\t',m.group(4))
```

```python
text = ['1 123 456 7890','+1 (123) 456 7890','123-456-7890','123.456.7890','1234567890']
pat = r'(?P<add1>\+?\d?)\s?(?P<area>\(?\d{3}\)?)[\W]?(?P<first3>\d{3})[\W]?(?P<last4>\d{4})'
pact = re.compile(pat)
for i in text:
    m = re.search(pact,i)
    if m:
        print(m.group(),'\t', m.group('add1'),'\t',m.group('area'),'\t',m.group('first3'),'\t',m.group('last4'))
```
